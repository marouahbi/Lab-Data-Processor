```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/macbookpro/Desktop/ultrasonic waves/12 slits.csv')
```

```python
df = df.drop([0,1,2,3])
df.head()
```

```python
df = df.drop('Time t / s', axis=1)
df = df.drop('Resistance R_A1 / kΩ', axis=1)
df = df.drop('Voltage U_B1 / V', axis=1)
df = df.drop('Relay R_1', axis=1)
df.head()
```

```python
df = df.reset_index()
```

```python
df = df.drop_duplicates(subset=['Angle f_1','Voltage U_B2 / V'],keep= 'first',ignore_index=True)
```

```python
df.info
```

```python
df = df.apply(pd.to_numeric, errors='coerce')
df.head(643)
```

```python
plt.plot(df['Angle f_1'], df["Voltage U_B2 / V"])
plt.xlabel("Angle")
plt.ylabel("Voltage")
plt.title("Diffraction patterns for 12 slits: voltage vs rotary stage angle")
plt.grid()
plt.xlim(-100, 100)
plt.show()
```
